# 网络的正向传播与反向更新

在神经网络理论中，构建了正向传播得到输出，计算loss反向更新神经元权值的模型体系。这是现代神经网络模型的理论基础，也是每一个初学者理解模型的第一步。假如使用公式来解释过程，未免会显得晦涩枯燥，这里举实例做计算，使过程更加容易被理解。

## 1.模型理解

在本节中，将以单层感知机模型作为演示，结构如下：  
![avatar](img/nn.png)
其中：  
i1，i2属于输入层  
h1，h2属于隐藏层  
o1，o2属于输出层  
w1，w2，w3...w8属于权重  
b1，b2属于偏置项  
激活函数采用Sigmoid函数  

## 2.前向传播
首先，我们先定义数据值：  
&emsp;&emsp;i1 = 0.05,&emsp;&emsp;i2 = 0.10  
&emsp;&emsp;o1 = 0.01,&emsp;&emsp;o2 = 0.99  
初始权重：  
&emsp;&emsp;w1 = 0.15,&emsp;&emsp;w2 = 0.20,&emsp;&emsp;w3 = 0.25,&emsp;&emsp;w4 = 0.30  
&emsp;&emsp;w5 = 0.40,&emsp;&emsp;w6 = 0.45,&emsp;&emsp;w7 = 0.50,&emsp;&emsp;w8 = 0.55  
我们期望输入数据i1，i2，使输出尽可能的靠近原始输出o1，o2。  
### 2.1 输入层===>隐藏层
计算神经元h1 的输入加权和：  
$h1=w_1*i_1+w_2*i_2+b_1$  
$h1=0.15*0.05+0.2*0.1+0.35*1=0.3775$  
使用激活函数做非线性化处理：  
$h1=sigmoid(0.3775)=0.593269992$  
同理，计算出神经元h2的输出o2：
$h2=0.596884378$
### 2.2 隐藏层===>输出层
计算输出层神经元o1和o2的值：  
$o1 = w_5*h1+w_6*h2+b_2$
$o1 = 0.4*0.593269992+0.45*0.596884378+0.6 = 1.105905967$
$out_{o1} = sigmoid(1.105905967) = 0.75136507$  
同理可得o2的输出为：  
$out_{o2} = 0.772928465$  
这样前向传播的过程就结束了，我们得到了输出值为[0.75136507,0.772928465],与真实值[0.01,0.99]相差还比较大，所以我们要计算真实值与预测值之间的误差进行反向传播，更新权值，重新计算输出。

## 3.反向传播
### 3.1计算总误差（Loss）
设定损失函数为欧氏距离：  
<center>$Loss=\sum\frac{1}{2}(target-output)^2$</center> 
现在有两个输出，所以要分别计算o1和o2的误差，总误差为两者之和：
<center>$L_{o1}=\sum\frac{1}{2}(target-output)^2=\frac{1}{2}(0.01-0.75136507)^2=0.274811083$</center> 
同理：
<center>$L_{o2}=0.023560026$</center> 
<center>$L_{total}=L_{o1}+L_{o2}=0.274811083+0.023560026=0.298371109$</center>
现在我们得到了误差值，下一步就需要对前面的权值进行更新。

### 3.2 权值更新 （隐含层===>输出层）
以权重参数w5为例，如果我们想知道w5对整体误差产生了多少影响，我们可以用整体误差对w5求偏导求出：(链式法则)
<font size=5><center>$\frac{\partial L_{total}}{\partial w_5}=\frac{\partial L_{total}}{\partial {out_{o1}}}*\frac{\partial {out_{o1}}}{\partial {o1}}*\frac{\partial {o1}}{\partial {w5}}$</center></font>
下面的图可以更直观的看清楚误差是怎样反向传播的：
![avatar](img/img1.png)
现在我们来计算每步式子的值

### 3.2.1 计算$\frac{\partial L_{total}}{\partial {out_{o1}}}$
首先我们已知$L_{total}$:  
<center>$L_{total}=\frac{1}{2}(target_{o1}-out_{o1})^2+\frac{1}{2}(target_{o2}-out_{o2})^2$</center>  
对$L_{total}$求$out_{o1}$的偏导：  
<center>$\frac{\partial L_{total}}{\partial out_{o1}}=2*\frac{1}{2}(target_{o1}-out_{o1})^{2-1}*-1+0$</center>
<center>$\frac{\partial L_{total}}{\partial out_{o1}}=-(target_{o1}-out_{o1})=-(0.01-0.75136507)=0.74136507$</center>
### 3.2.2 计算$\frac{\partial {out_{o1}}}{\partial {o1}}$
此处的$out_{o1}$实际上就是使用激活函数非线性化o1：  
$out_{o1}=sigmoid(o1)=\frac1{1+exp(o1)}$
对其求o1的偏导：  
$\frac{\partial out_{o1}}{\partial o1}=out_{o1}(1-out_{o1})=0.75136507(1-0.75136507)=0.186815602$
### 3.2.3 计算$\frac{\partial {o1}}{\partial {w5}}$
现在我们已知：  
$o1=w_5*out_{h1}+w_6*out_{h2}+b_2$  
对其求w5的偏导：  
$\frac{\partial {o1}}{\partial {w5}}=1*out_{h1}*w_5^{1-1}+0+0=out_{h1}=0.593269992$
### 3.2.4 更新参数
根据计算得到的数值带入公式：  
<font size=5><center>$\frac{\partial L_{total}}{\partial w_5}=\frac{\partial L_{total}}{\partial {out_{o1}}}*\frac{\partial {out_{o1}}}{\partial {o1}}*\frac{\partial {o1}}{\partial {w5}}$</center></font>
<center>$\frac{\partial L_{total}}{\partial w_5}=0.74136507*0.186815602*0.593269992=0.082167041$</center>  
这里引入学习率的概念，即每一次更新参数的量：  
学习率:$\eta=0.5$  
<font size=5>$w_5^+=w_5-\eta*\frac{\partial L_{total}}{\partial w_5}=0.4-0.5*0.082167041=0.35891648$</font>  
同理可得：  
<font size=5>$w_6^+=0.408666186$  
$w_7^+=0.511301270$  
$w_8^+=0.561370121$ </font>   
可以自己练习手推以下这里的三个参数。

### 3.3 权值更新（隐含层===>隐含层）
方法与上一层相似，但是有部分地方需要更改。  
在上文中计算总误差对w5的偏导是从：$out_{o1}---->o_1---->w_5$  
但是在隐含层之间的权值更新时，是：$out_{h1}---->h_1---->w_1$  
其中$out_{h1}$会接受从$L_{o1},L_{o2}$两个地方传来的误差，所以这两个地方都需要计算。  
<font size=5><center>$\frac{\partial L_{total}}{\partial w_1}=\frac{\partial L_{total}}{\partial {out_{h1}}}*\frac{\partial {out_{h1}}}{\partial {h1}}*\frac{\partial {h1}}{\partial {w1}}$</center></font>  
其中：  
<font size=5><center>$\frac{\partial L_{total}}{\partial {out_{h1}}}=\frac{\partial {L_{o1}}}{\partial out_{h1}}*\frac{\partial L_{o2}}{\partial out_{h1}}$</center></font> 
![avatar](img/nn.png)

### 3.3.1 计算 $\frac{\partial {L_{o1}}}{\partial out_{h1}}$
我们并不能直接对o1的损失值求$out_{h1}$的偏导，所以仍然要用到链式法则：  
<center>$\frac{\partial {L_{o1}}}{\partial out_{h1}}=\frac{\partial {L_{o1}}}{\partial {o1}}*\frac{\partial {o1}}{\partial out_{h1}}$</center>
其中，$\frac{\partial {L_{o1}}}{\partial {o1}}$我们同样不能直接得到，还要使用链式法则：
<center>$\frac{\partial {L_{o1}}}{\partial {o1}}=\frac{\partial {L_{o1}}}{\partial out_{o1}}*\frac{\partial out_{o1}}{\partial {o1}}=0.74136507*0.186815602=0.138498562$</center>  
得到前半部分的值后，再看后半部分，已知其中o1：  
<center>$o_1=w_5*out_{h1}+w_6*out_{h2}+b_2$</center>   
对其求$out_{h1}$的偏导就为：  
<center>$\frac{\partial {o1}}{\partial out_{h1}}=w_5=0.40$</center>  
带入前面的式子：  
<center>$\frac{\partial {L_{o1}}}{\partial out_{h1}}=\frac{\partial {L_{o1}}}{\partial {o1}}*\frac{\partial {o1}}{\partial out_{h1}}=0.138498562*0.40=0.055399425$</center>   
同理，可以计算出：  
<center>$\frac{\partial {L_{o2}}}{\partial out_{h1}}=-0.019049119$</center>  
那么就可以得到：  
<center>$\frac{\partial L_{total}}{\partial {out_{h1}}}=\frac{\partial {L_{o1}}}{\partial out_{h1}}*\frac{\partial L_{o2}}{\partial out_{h1}}=0.055399425-0.019049119=0.036350306$</center>   


### 3.3.2 计算 $\frac{\partial {out_{h1}}}{\partial {h1}}$  
此处可以参考3.2.2部分：  
<center>$\frac{\partial out_{h1}}{\partial h1}=out_{h1}(1-out_{h1})=0.59326999(1-0.59326999)=0.241300709$</center>  
### 3.3.3 计算 $\frac{\partial {h1}}{\partial {w1}}$  
这部分也较为简单：  
<center>$h_1=w_1*i_1+w_2*i_2+b_1$   
$\frac{\partial {h_1}}{\partial {w_1}}=i_1=0.05$</center>   
### 3.3.4 权值更新    
<center>$\frac{\partial L_{total}}{\partial w_1}=\frac{\partial L_{total}}{\partial {out_{h1}}}*\frac{\partial {out_{h1}}}{\partial {h1}}*\frac{\partial {h1}}{\partial {w1}}$  
$=0.036350306*0.241300709*0.05=0.000438568$</center>  
接下来同3.2.4相同：  
定义学习率$\eta=0.5$,则：  
<>$w_1^+=w_1-\eta*\frac{\partial L_{total}}{\partial w_1}=0.15-0.5*0.000438568=0.149780716$  
同理，可更新w2,w3,w4的权值：  
$w_2^+=0.19956143$  
$w_3^+=0.24975114$  
$w_4^+=0.29950299$  

```python
#coding:utf-8
import random
import math

#
#   参数解释：
#   "pd_" ：偏导的前缀
#   "d_" ：导数的前缀
#   "w_ho" ：隐含层到输出层的权重系数索引
#   "w_ih" ：输入层到隐含层的权重系数的索引

class NeuralNetwork:
    LEARNING_RATE = 0.5

    def __init__(self, num_inputs, num_hidden, num_outputs, hidden_layer_weights = None, hidden_layer_bias = None, output_layer_weights = None, output_layer_bias = None):
        self.num_inputs = num_inputs

        self.hidden_layer = NeuronLayer(num_hidden, hidden_layer_bias)
        self.output_layer = NeuronLayer(num_outputs, output_layer_bias)

        self.init_weights_from_inputs_to_hidden_layer_neurons(hidden_layer_weights)
        self.init_weights_from_hidden_layer_neurons_to_output_layer_neurons(output_layer_weights)

    def init_weights_from_inputs_to_hidden_layer_neurons(self, hidden_layer_weights):
        weight_num = 0
        for h in range(len(self.hidden_layer.neurons)):
            for i in range(self.num_inputs):
                if not hidden_layer_weights:
                    self.hidden_layer.neurons[h].weights.append(random.random())
                else:
                    self.hidden_layer.neurons[h].weights.append(hidden_layer_weights[weight_num])
                weight_num += 1

    def init_weights_from_hidden_layer_neurons_to_output_layer_neurons(self, output_layer_weights):
        weight_num = 0
        for o in range(len(self.output_layer.neurons)):
            for h in range(len(self.hidden_layer.neurons)):
                if not output_layer_weights:
                    self.output_layer.neurons[o].weights.append(random.random())
                else:
                    self.output_layer.neurons[o].weights.append(output_layer_weights[weight_num])
                weight_num += 1

    def inspect(self):
        print('------')
        print('* Inputs: {}'.format(self.num_inputs))
        print('------')
        print('Hidden Layer')
        self.hidden_layer.inspect()
        print('------')
        print('* Output Layer')
        self.output_layer.inspect()
        print('------')

    def feed_forward(self, inputs):
        hidden_layer_outputs = self.hidden_layer.feed_forward(inputs)
        return self.output_layer.feed_forward(hidden_layer_outputs)

    def train(self, training_inputs, training_outputs):
        self.feed_forward(training_inputs)

        # 1. 输出神经元的值
        pd_errors_wrt_output_neuron_total_net_input = [0] * len(self.output_layer.neurons)
        for o in range(len(self.output_layer.neurons)):

            # ∂E/∂zⱼ
            pd_errors_wrt_output_neuron_total_net_input[o] = self.output_layer.neurons[o].calculate_pd_error_wrt_total_net_input(training_outputs[o])

        # 2. 隐含层神经元的值
        pd_errors_wrt_hidden_neuron_total_net_input = [0] * len(self.hidden_layer.neurons)
        for h in range(len(self.hidden_layer.neurons)):

            # dE/dyⱼ = Σ ∂E/∂zⱼ * ∂z/∂yⱼ = Σ ∂E/∂zⱼ * wᵢⱼ
            d_error_wrt_hidden_neuron_output = 0
            for o in range(len(self.output_layer.neurons)):
                d_error_wrt_hidden_neuron_output += pd_errors_wrt_output_neuron_total_net_input[o] * self.output_layer.neurons[o].weights[h]

            # ∂E/∂zⱼ = dE/dyⱼ * ∂zⱼ/∂
            pd_errors_wrt_hidden_neuron_total_net_input[h] = d_error_wrt_hidden_neuron_output * self.hidden_layer.neurons[h].calculate_pd_total_net_input_wrt_input()

        # 3. 更新输出层权重系数
        for o in range(len(self.output_layer.neurons)):
            for w_ho in range(len(self.output_layer.neurons[o].weights)):

                # ∂Eⱼ/∂wᵢⱼ = ∂E/∂zⱼ * ∂zⱼ/∂wᵢⱼ
                pd_error_wrt_weight = pd_errors_wrt_output_neuron_total_net_input[o] * self.output_layer.neurons[o].calculate_pd_total_net_input_wrt_weight(w_ho)

                # Δw = α * ∂Eⱼ/∂wᵢ
                self.output_layer.neurons[o].weights[w_ho] -= self.LEARNING_RATE * pd_error_wrt_weight

        # 4. 更新隐含层的权重系数
        for h in range(len(self.hidden_layer.neurons)):
            for w_ih in range(len(self.hidden_layer.neurons[h].weights)):

                # ∂Eⱼ/∂wᵢ = ∂E/∂zⱼ * ∂zⱼ/∂wᵢ
                pd_error_wrt_weight = pd_errors_wrt_hidden_neuron_total_net_input[h] * self.hidden_layer.neurons[h].calculate_pd_total_net_input_wrt_weight(w_ih)

                # Δw = α * ∂Eⱼ/∂wᵢ
                self.hidden_layer.neurons[h].weights[w_ih] -= self.LEARNING_RATE * pd_error_wrt_weight

    def calculate_total_error(self, training_sets):
        total_error = 0
        for t in range(len(training_sets)):
            training_inputs, training_outputs = training_sets[t]
            self.feed_forward(training_inputs)
            for o in range(len(training_outputs)):
                total_error += self.output_layer.neurons[o].calculate_error(training_outputs[o])
        return total_error

class NeuronLayer:
    def __init__(self, num_neurons, bias):

        # 同一层的神经元共享一个截距项b
        self.bias = bias if bias else random.random()

        self.neurons = []
        for i in range(num_neurons):
            self.neurons.append(Neuron(self.bias))

    def inspect(self):
        print('Neurons:', len(self.neurons))
        for n in range(len(self.neurons)):
            print(' Neuron', n)
            for w in range(len(self.neurons[n].weights)):
                print('  Weight:', self.neurons[n].weights[w])
            print('  Bias:', self.bias)

    def feed_forward(self, inputs):
        outputs = []
        for neuron in self.neurons:
            outputs.append(neuron.calculate_output(inputs))
        return outputs

    def get_outputs(self):
        outputs = []
        for neuron in self.neurons:
            outputs.append(neuron.output)
        return outputs

class Neuron:
    def __init__(self, bias):
        self.bias = bias
        self.weights = []

    def calculate_output(self, inputs):
        self.inputs = inputs
        self.output = self.squash(self.calculate_total_net_input())
        return self.output

    def calculate_total_net_input(self):
        total = 0
        for i in range(len(self.inputs)):
            total += self.inputs[i] * self.weights[i]
        return total + self.bias

    # 激活函数sigmoid
    def squash(self, total_net_input):
        return 1 / (1 + math.exp(-total_net_input))


    def calculate_pd_error_wrt_total_net_input(self, target_output):
        return self.calculate_pd_error_wrt_output(target_output) * self.calculate_pd_total_net_input_wrt_input();

    # 每一个神经元的误差是由平方差公式计算的
    def calculate_error(self, target_output):
        print("真实值：",target_output,"预测值:",self.output)
        return 0.5 * (target_output - self.output) ** 2


    def calculate_pd_error_wrt_output(self, target_output):
        return -(target_output - self.output)


    def calculate_pd_total_net_input_wrt_input(self):
        return self.output * (1 - self.output)


    def calculate_pd_total_net_input_wrt_weight(self, index):
        return self.inputs[index]


# 文中的例子:

nn = NeuralNetwork(2, 2, 2, hidden_layer_weights=[0.15, 0.2, 0.25, 0.3], hidden_layer_bias=0.35, output_layer_weights=[0.4, 0.45, 0.5, 0.55], output_layer_bias=0.6)
for step in range(10000):
  nn.train([0.05, 0.1], [0.01, 0.99])
  print("setp:", step, "total Loss:", round(nn.calculate_total_error([[[0.05, 0.1], [0.01, 0.99]]]), 9))

```
